Here's your chance to write a function that will be useful for the Advanced Lane-Finding Project at the end of this lesson! Your goal in this exercise is to identify pixels where the gradient of an image falls within a specified threshold range.

## Example


We call the function definition defined in the [apply_sobel.py](apply_sobel.py)

```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
```

Import funntion implemented from appyl_sobel.py

```python
import os
import sys
sys.path.append(os.path.abspath("apply_sobel.py"))

from apply_sobel import * 
```

Read in an image and grayscale it

```python
image = mpimg.imread('test_images/signs_vehicles_xygrad.png')
```

Run the function

```python
grad_binary = abs_sobel_thresh(image, orient='y', thresh_min=30, thresh_max=100)
```

Plot the result

```python
f, (ax1, ax2) = plt.subplots(1, 2, figsize=(24, 9))
f.tight_layout()
ax1.imshow(image)
ax1.set_title('Original Image', fontsize=50)
ax2.imshow(grad_binary, cmap='gray')
ax2.set_title('Thresholded Gradient', fontsize=50)
plt.subplots_adjust(left=0., right=1, top=0.9, bottom=0.)
```
