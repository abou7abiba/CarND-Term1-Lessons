In this exercise, you can try applying a color threshold in the HLS color space. If you have any problems with the implementation, please refer to the previous video and instructor notes.

Your task here is to write a function that takes in an image and threshold values and returns a binary output from applying the threshold to the S-channel. 

## Example


We call the function definition defined in the [hls-threshold.py](hls-threshold.py)

```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg

import cv2
```

Import funntion implemented from appyl_sobel.py

```python
import os
import sys
sys.path.append(os.path.abspath("hls_threshold.py"))

from hls_threshold import * 
```

Read in an image and grayscale it

```python
image = mpimg.imread('test_images/test6.jpg')
hls = cv2.cvtColor(image, cv2.COLOR_RGB2HLS)
H = hls[:,:,0]
L = hls[:,:,1]
S = hls[:,:,2]
```

Run the function

```python
hls_binary = hls_select(image, thresh=(90, 255))
```

Plot the result

```python
f, (ax1, ax2, ax3) = plt.subplots(1, 3, figsize=(24, 9))
f.tight_layout()
ax1.imshow(image)
ax1.set_title('Original Image', fontsize=50)
ax2.imshow(S)
ax2.set_title('S IMAGE', fontsize=50)
ax3.imshow(hls_binary, cmap='gray')
ax3.set_title('Thresholded S', fontsize=50)
plt.subplots_adjust(left=0., right=1, top=0.9, bottom=0.)
```

```python
f, (ax1, ax2, ax3) = plt.subplots(1, 3, figsize=(24, 9))
f.tight_layout()
ax1.imshow(H)
ax1.set_title('H Image', fontsize=50)
ax2.imshow(L)
ax2.set_title('L Image', fontsize=50)
ax3.imshow(S)
ax3.set_title('S IMAGE', fontsize=50)
plt.subplots_adjust(left=0., right=1, top=0.9, bottom=0.)
```
