# Find the lines using the previous polynomial

We use the polynomial of the previous fram to detect the starting point of the search. The search will be aroung this polynomial +/- a given margin to search in it. This margin indicated by the green area in the image below.



We call the function definition defined in the [poly_based_search_quiz.py](poly_based_search_quiz.py)

```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
```

Import funntion implemented from [poly_based_search_quiz.py](poly_based_search_quiz.py)

```python
import os
import sys
sys.path.append(os.path.abspath("poly_based_search_quiz.py"))

from poly_based_search_quiz import * 
```

Read in an image and grayscale it

```python
binary_warped = mpimg.imread('test_images/warped-example.jpg')
```

Run the function

```python
result = search_around_poly(binary_warped)
```

Plot the result

```python
f, (ax1, ax2) = plt.subplots(1, 2, figsize=(24, 9))
f.tight_layout()
ax1.imshow(binary_warped, cmap='gray')
ax1.set_title('Original Image', fontsize=50)
ax2.imshow(result)
ax2.set_title('Showing Search Area', fontsize=50)
plt.subplots_adjust(left=0., right=1, top=0.9, bottom=0.)
```
